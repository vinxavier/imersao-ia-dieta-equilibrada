# Dieta Equílibrada

O principal objetivo é retornar ao usuário uma dieta equilíbrada de proteínas para uma rotina mais saudável específica para o seu peso.

A recomendaçoes aqui feitas são geradas por uma IA e deve apenas servir de material de apoio para um nutricionista.

Como padrão será indica 2g de proteínas por dia por kilograma da pesso, seguindo essa matéria do metropoles: https://www.metropoles.com/colunas/claudia-meireles/qual-quantidade-de-proteina-devo-consumir-diariamente-expert-responde

```python
!pip install -q -U google-generativeai
```

```python
import google.generativeai as genai
from google.colab import userdata
from IPython.display import display
from IPython.display import Markdown
```

```python
GOOGLE_API_KEY=userdata.get('api_key')

genai.configure(api_key=GOOGLE_API_KEY)

model = genai.GenerativeModel('gemini-1.5-pro-latest')
```

### Sua vez!

Digite abaixo o nome do seu paciente.

Opcionalmente digite a quantidade de proteína ideal por kilo do seu paciente. O padrão é 2g, apenas dê enter se quiser pular.

Digite também o peso do seu paciente.

```python
paciente = input("Qual o nome do seu paciente? ")
proteina_base = int(input("Qual quantidade de proteína ideal por kilo do seu paciente? ") or 2)
peso_paciente = float(input("Qual o peso do seu paciente em kg? "))
```

```python
prompt = f"""Sabendo que um individuo precisa ingerir {proteina_base}g de proteínas por dia para cada kg de seu peso, e que cada pessoa tem 4 a 5 refeições por dia.

Se um individuo pesa {peso_paciente}kg, gere a quantidade de proteína por dia que esse individuo precisa ingerir.

Além disso, sugira alimentos com proteínas e quantidade desse alimento que esse sujeito deve ingerir em cada refeição do dia para cada dia da semana, gerando um cardapio diferente em cada dia. Defina com exatidão a quantidade dos alimentos propostos nas refeições para que atinja o minimo de proteínas necessarias por dia

Gere no final uma lista de compras para a semana com a quantidade total de cada alimento que esse individuo precisa comprar para preencher toda a lista gerada.

Return the response in a beutifull structured markdown."""

response = model.generate_content(prompt)

display(Markdown(response.text))
```

# Gerar PDF para impressão da dieta

Abaixo vamos pegar o nosso resultado e gerar um PDF para ser impresso que será salvo nos arquivos desse colab.

```python
!pip install markdown-pdf
```

```python
from markdown_pdf import MarkdownPdf, Section

pdf = MarkdownPdf(toc_level=0)
pdf.add_section(Section(response.text, paper_size="A4-L"))
pdf.meta["title"] = f"Dieta para {paciente}"
pdf.meta["author"] = "Gerado por Dieta Equilíbrada"

pdf.save(f"Dieta {paciente}.pdf")
```
